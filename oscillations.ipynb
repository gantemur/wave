```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

$$
\begin{cases}x''+2kx'+\alpha^2 x=0\\y''+2ky'+\beta^2 y=0\end{cases}
$$
тэгшитгэлээр өгөгдөх 2 хэмжээстэй унтрах гармоник хэлбэлзэл авч үзье. Энэ хөдөлгөөнийг $x(t)=e^{-kt}\xi(t)$, $y(t)=e^{-kt}\eta(t)$ гэсэн орлуулгаар гармоник хэлбэлзэлд шилжүүлж болох ба эцсийн шийд нь
$$
\begin{cases}
x(t)=Ae^{-kt}\cos(\omega_1t+a)\\
y(t)=Be^{-kt}\sin(\omega_2t+b)
\end{cases}
$$
гарна. Үүнд $A,a,B,b$ нь анхны нөхцлөөс хамаарах тогтмол тоонууд ба $\omega_1=\sqrt{\alpha^2-k^2}$, $\omega_2=\sqrt{\beta^2-k^2}$.

```python
omx=1
omy=1.1
k=.01
A=1
B=1
a=0
b=3
tlist=np.linspace(0, 200, 5000)
x=[A*np.exp(-k*t)*np.cos(omx*t+a) for t in tlist]
y=[B*np.exp(-k*t)*np.sin(omy*t+b) for t in tlist]
plt.plot(x,y)
plt.show()
```
